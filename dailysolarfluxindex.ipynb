```python
import matplotlib.pyplot as plt


from matplotlib import font_manager

import matplotlib.dates
import matplotlib.ticker
import pandas as pd

# Create an array of solar flux index values.
SFI = [146,136,142,140,126,153,159,167,176,176,186,]

# Create a DataFrame.
df = pd.DataFrame({'date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05','2024-01-06','2024-01-07','2024-01-08','2024-01-09','2024-01-10','2024-01-11'],
                   'solar_flux_index': SFI})

# Convert the date column to a datetime object.
df['date'] = pd.to_datetime(df['date'])

# Create a plot.
plt.plot(df['date'], df['solar_flux_index'])
#plt.scatter(df['date'], df['solar_flux_index'])

# Set the x-axis tick labels to only display the month and year.
#plt.xticks(rotation=45, ha='right', minor=False)
plt.xticks(df['date'], rotation=45, ha='right')
# plt.set_major_locator(matplotlib.dates.MonthLocator(interval=1))
# plt.set_major_formatter(matplotlib.dates.DateFormatter('%b %Y'))

# Show the plot.
plt.ylabel('Solar Flux Index ')
plt.title('Solar Flux Index\n Starting from 1st Jan 2024')

for i in range(len(df)):
    plt.annotate(df['solar_flux_index'][i], (df['date'][i], df['solar_flux_index'][i]))

#plt.set_title_font(font_manager.FontProperties(family='Arial'))
#plt.set_title_font('Arial')
#plt.set_xlabel_font('Helvetica')

plt.plot(df['date'], df['solar_flux_index'])

fig = plt.gcf()
fig.set_facecolor('yellow')



plt.show()
```
